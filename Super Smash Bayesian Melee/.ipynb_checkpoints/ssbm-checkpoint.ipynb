# Analyse bayesienne des 4 meilleurs joueurs de Super Smash Bros. Melee

```python
import re
import numpy as np
from scipy import stats
import pymc3 as pm
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
```

```python
def get_player_results(player):
    with open(f"players/{player}.txt", "r") as f:
        text = f.read()
        text = text.replace("\n", " ")
        matchs = re.findall(r"(\d+) : (\d+)", text)
        return np.array([int(int(v) > int(d)) for v, d in matchs])
```

## Récupération des données et caractéristiques des joueurs

```python
hungrybox = get_player_results("hungrybox")
plup = get_player_results("plup")
armada = get_player_results("armada")
mew2king = get_player_results("mew2king")
```

## Définition des "prior belief" du modèle et sampling

```python
with pm.Model() as model:
#     A_p = pm.Beta("A_p", armada.sum() + 1, (1 - armada).sum() + 1)
#     H_p = pm.Beta("H_p", hungrybox.sum() + 1, (1 - hungrybox).sum() + 1)
#     M_p = pm.Beta("M_p", mew2king.sum() + 1, (1 - mew2king).sum() + 1)
#     P_p = pm.Beta("P_p", plup.sum() + 1, (1 - plup).sum() + 1)

#     A_p = pm.Beta("A_p", 12, 12)
#     H_p = pm.Beta("H_p", 12, 12)
#     M_p = pm.Beta("M_p", 12, 12)
#     P_p = pm.Beta("P_p", 12, 12)

    A_p = pm.Uniform("A_p", 0, 1)
    H_p = pm.Uniform("H_p", 0, 1)
    M_p = pm.Uniform("M_p", 0, 1)
    P_p = pm.Uniform("P_p", 0, 1)
    
    A = pm.Bernoulli("A", p=A_p, observed=armada)
    H = pm.Bernoulli("H", p=H_p, observed=hungrybox)
    M = pm.Bernoulli("M", p=M_p, observed=mew2king)
    P = pm.Bernoulli("P", p=P_p, observed=plup)
    
    
    AvH = pm.Deterministic("AvH", A_p - H_p)
    AvM = pm.Deterministic("AvM", A_p - M_p)
    AvP = pm.Deterministic("AvP", A_p - P_p)
    
    HvM = pm.Deterministic("HvM", H_p - M_p)
    HvP = pm.Deterministic("HvP", H_p - P_p)
    
    MvP = pm.Deterministic("MvP", M_p - P_p)
    
    start = pm.find_MAP()
    trace = pm.sample(20000, tune=5000, start=start)
    
    burned_trace = trace[10000::2]
```

## Visualisation des "posterior distributions"

```python
pm.traceplot(burned_trace);
```

```python
pm.plot_posterior(burned_trace);
```

## Résultats

```python
print("Probability that Armada is better than Hungrybox : ", (burned_trace["AvH"] > 0).mean())
print("Probability that Armada is better than Mew2king : ", (burned_trace["AvM"] > 0).mean())
print("Probability that Armada is better than Plup : ", (burned_trace["AvP"] > 0).mean())
print("Probability that Hungrybox is better than Mew2king : ", (burned_trace["HvM"] > 0).mean())
print("Probability that Hungrybox is better than Plup : ", (burned_trace["HvP"] > 0).mean())
print("Probability that Mew2king is better than Plup : ", (burned_trace["MvP"] > 0).mean())
```

## Experimentation avec un joueur hypothétique

A chaque partie, le joueur hypothétique lance une pièce. Si elle tombe sur face, il joue avec une probabilité 0.8 de remporter la partie (bien meilleur que n'importe quel vrai joueur dont Armada). Si elle tombe sur pile il gagne avec une probabilité 0.25, sinon il perd. 

Comment modéliser ces contraintes et comparer ce faux joueur au meilleur vrai joueur (Armada) ?

```python
fake = np.array([
    x 
    if np.random.random() > 0.5 
    else np.random.random() > 0.75
    for x in stats.bernoulli(0.8).rvs(1000)
])

stats.describe(fake)
```

### Modélisation sans prise en compte des contraintes imposées au joueur

```python
with pm.Model() as model:
    A_p = pm.Uniform("A_p", 0, 1)
    F_p = pm.Uniform("F_p", 0, 1)
    
    A = pm.Bernoulli("A", p=A_p, observed=armada)
    F = pm.Bernoulli("F", p=F_p, observed=fake)
    
    
    AvF = pm.Deterministic("AvF", A_p - F_p)

    start = pm.find_MAP()
    trace = pm.sample(20000, tune=5000, start=start)
    
    burned_trace = trace[10000::2]
```

```python
pm.traceplot(burned_trace);
pm.plot_posterior(burned_trace);
print("Probability that Armada is better than Fake : ", (burned_trace["AvF"] > 0).mean())
```

### Modélisation avec prise en compte des contraintes imposées au joueur

```python
with pm.Model() as model:
    A_p = pm.Uniform("A_p", 0, 1)
    F_p = pm.Uniform("F_p", 0, 1)
    
    F_p_skewed = pm.Deterministic("F_p_skewed", 0.5 * F_p + 0.5 * 0.25)
    
    A = pm.Bernoulli("A", p=A_p, observed=armada)
    F = pm.Bernoulli("F", p=F_p_skewed, observed=fake)
    
    
    AvF = pm.Deterministic("AvF", A_p - F_p)

    start = pm.find_MAP()
    trace = pm.sample(20000, tune=5000, start=start)
    
    burned_trace = trace[10000::2]
```

```python
pm.traceplot(burned_trace);
pm.plot_posterior(burned_trace);
print("Probability that Armada is better than Fake : ", (burned_trace["AvF"] > 0).mean())
```
